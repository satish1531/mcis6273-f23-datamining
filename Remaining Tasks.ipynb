```python
# Task 1: Remove Stopwords and Create all_ns_words.csv
import pandas as pd

stopwords_file_path = "stopwords-en.txt"
# Read the stopwords from the local file using pandas
stopwords = pd.read_csv(stopwords_file_path, header=None, names=["word"])

# Read your existing CSV file (all_words.csv)
all_words_df = pd.read_csv("all_words.csv")

# Remove stopwords from the all_words dataframe
all_ns_words_df = all_words_df[~all_words_df["word"].isin(stopwords["word"])]

# Calculate the probability for each word
total_word_count = all_ns_words_df["frequency"].sum()
all_ns_words_df["probability"] = all_ns_words_df["frequency"] / total_word_count

# Export the result to a new CSV file (all_ns_words.csv)
all_ns_words_df.to_csv("all_ns_words.csv", index=False)

# Task 3: Answering questions
# Read the updated CSV file with probabilities
all_ns_words_df = pd.read_csv("all_ns_words.csv")

# 1. Number of unique non-stop words
unique_non_stopwords = all_ns_words_df["word"].nunique()
print("1. Number of unique non-stop words:", unique_non_stopwords)


# Least probable word(s)
min_probability = all_ns_words_df["probability"].min()
least_probable_words = all_ns_words_df[all_ns_words_df["probability"] == min_probability]["word"].tolist()
print("2. Least probable word(s):", least_probable_words)

# 3. Observations about probabilities
# Observations about probabilities
probability_mean = all_ns_words_df["probability"].mean()
probability_median = all_ns_words_df["probability"].median()
probability_variance = all_ns_words_df["probability"].var()
probability_range = all_ns_words_df["probability"].max() - all_ns_words_df["probability"].min()
probability_skewness = all_ns_words_df["probability"].skew()

print("3. Observations about probabilities:")
print("   - Mean Probability:", probability_mean)
print("   - Median Probability:", probability_median)
print("   - Variance of Probabilities:", probability_variance)
print("   - Range of Probabilities:", probability_range)
print("   - Skewness of Probabilities:", probability_skewness)

# 4. Comparing sentence probabilities
# Calculate the sum of probabilities for each word in the sentence
sentence_a = "If a belief is true, it can be deduced it is universal."
sentence_b = "Criticism of knowledge is counter to scientific results."

def calculate_sentence_probability(sentence, df):
    words = sentence.split()
    probability_sum = df[df["word"].isin(words)]["probability"].sum()
    return probability_sum

# Calculate probabilities for both sentences
sentence_a = "If a belief is true, it can be deduced it is universal."
sentence_b = "Criticism of knowledge is counter to scientific results."

probability_a = calculate_sentence_probability(sentence_a, all_ns_words_df)
probability_b = calculate_sentence_probability(sentence_b, all_ns_words_df)

# Print the calculated probabilities for both sentences
print("4. Comparing sentence probabilities:")
print("   - Probability for sentence A:", probability_a)
print("   - Probability for sentence B:", probability_b)

# Determine which sentence is more likely
if probability_a > probability_b:
    print("   Sentence A is more likely.")
elif probability_b > probability_a:
    print("   Sentence B is more likely.")
else:
    print("   Both sentences have the same probability.")
```
